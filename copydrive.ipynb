#**Files**

##**File Operation**

```
#@title Rename file/directory
directory = '' #@param {type:"string"}
source = '' #@param {type:"string"}
target = '' #@param {type:"string"}
%cd "$directory"
%mv "$source" "$target"
print('\33[34m'+'Rename complete.'+'\33[0m')
```

```
#@title Move file/directory
from datetime import datetime
now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")

source = '' #@param {type:"string"}
target = '' #@param {type:"string"}

print("\33[32m"+"Runtime started at "+date_now+"\33[0m")
print('\33[93m'+'Initializing..'+'\33[0m')
print('\33[34m'+'Moving..'+'\33[0m')

%mv "$source" "$target"

print('\33[34m'+'Complete.'+'\33[0m')

now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")
print("\33[32m"+"Runtime ended at "+date_now+"\33[0m")
```

```
#@title Copy file/directory
from subprocess import call
from datetime import datetime
now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")

source = '' #@param {type:"string"}
target = '' #@param {type:"string"}

def cp_dir(source, target):
    call(['cp', '-a', source, target]) 

print("\33[32m"+"Runtime started at "+date_now+"\33[0m")
print('\33[93m'+'Initializing..'+'\33[0m')
print('\33[34m'+'Copying..'+'\33[0m')

cp_dir(source, target)

print('\33[34m'+'Copy Complete, but make sure your file is safely copied..'+'\33[0m')

now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")
print("\33[32m"+"Runtime ended at "+date_now+"\33[0m")
```

```
#@title Delete file/directory
target = '' #@param {type:"string"}
#@markdown **(?)** use with caution, this will delete your *file/directory* permanently
print('\33[34m'+'Deleting..'+'\33[0m')
%rm -r "$target"
print('\33[34m'+'Done.'+'\33[0m')
```
